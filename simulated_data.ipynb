```python
import pandas as pd
import numpy as np

np.random.seed(42)
```

```python
# Caravana AZ approximate center
center_lat, center_lng = 33.4294, -111.9381

# Generate 15 fake delivery points within 0.1 degree lat/lng (~11 km)
num_deliveries = 15
latitudes = center_lat + np.random.uniform(-0.05, 0.05, num_deliveries)
longitudes = center_lng + np.random.uniform(-0.05, 0.05, num_deliveries)
```

```python
depot = {'location_id': 'Depot', 'lat': center_lat, 'lng': center_lng}
data = pd.DataFrame({
    'location_id': [f'Delivery_{i+1}' for i in range(num_deliveries)],
    'lat': latitudes,
    'lng': longitudes
})

data = pd.concat([pd.DataFrame([depot]), data], ignore_index=True)

data
```

```python
data.to_csv("carvana_delivery_locations.csv", index=False)
data.head()
```
